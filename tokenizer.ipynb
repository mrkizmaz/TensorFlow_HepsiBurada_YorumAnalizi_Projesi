```python
import tensorflow as tf
```

```python
import pandas as pd
```

```python
df = pd.read_csv('HB_Yorumlar/hb_yorumlar1.csv')
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer(
    num_words=None,
    filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
    lower=True,
    split=' ',
    char_level=False,
    oov_token=None,
    document_count=0)
```

```python
yorum = df['yorum'].values.tolist()
```

```python
yorum[0:10]
```

```python
tokenizer.fit_on_texts(yorum)
```

```python
from itertools import islice

def take(n, iterable):
    return islice(iterable, n)
```

```python
type(tokenizer.word_index.items())
```

```python
n_items = take(10000, tokenizer.word_index.items())
```
